```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import rcParams
```

```python
DenseNet121    = [0.79, 0.86, 0.81]
DenseNet169    = [0.78, 0.85, 0.82]
DenseNet201    = [0.76, 0.85, 0.8]
EfficientNetB2 = [0.64, 0.7, 0.55]
EfficientNetB4 = [0.71, 0.78, 0.65]
EfficientNetB6 = [0.71, 0.77, 0.69]
ResNet50       = [0.78, 0.81, 0.77]
ResNet101      = [0.53, 0.55, 0.53]
ResNet152      = [0.69, 0.69, 0.58]
Multibranch    = [0.83, 0.81, 0.92]
```

```python
np.array([DenseNet121,DenseNet169,DenseNet201,
          EfficientNetB2,EfficientNetB4,EfficientNetB6,
          ResNet50,ResNet101,ResNet152,
          Multibranch])
```

```python
np.array(Multibranch)
```

```python
data1 = pd.DataFrame([DenseNet121], columns=['Benign','Malignant','Normal']).assign(Models='DenseNet121')
data2 = pd.DataFrame([DenseNet169], columns=['Benign','Malignant','Normal']).assign(Models='DenseNet169')
data3 = pd.DataFrame([DenseNet201], columns=['Benign','Malignant','Normal']).assign(Models='DenseNet201')
data4 = pd.DataFrame([EfficientNetB2], columns=['Benign','Malignant','Normal']).assign(Models='EfficientNetB2')
data5 = pd.DataFrame([EfficientNetB4], columns=['Benign','Malignant','Normal']).assign(Models='EfficientNetB4')
data6 = pd.DataFrame([EfficientNetB6], columns=['Benign','Malignant','Normal']).assign(Models='EfficientNetB6')
data7 = pd.DataFrame([ResNet50], columns=['Benign','Malignant','Normal']).assign(Models='ResNet50')
data8 = pd.DataFrame([ResNet101], columns=['Benign','Malignant','Normal']).assign(Models='ResNet101')
data9 = pd.DataFrame([ResNet152], columns=['Benign','Malignant','Normal']).assign(Models='ResNet152')
data10 = pd.DataFrame([Multibranch], columns=['Benign','Malignant','Normal']).assign(Models='MultiBranch UNet')
cdf = pd.concat([data1, data2, data3,data4,data5,data6,data7,data8,data9,data10])
cdf
```

```python
mdf = pd.melt(cdf, id_vars=['Models'], var_name=['Tumor'])
mdf
```

```python
rcParams['figure.figsize'] = 25,10
sns.set(font_scale =2)
ax = sns.barplot(x="Models", y="value", hue="Tumor", data=mdf, errwidth=0.9) 
ax.set(ylabel='Dice Score')
# ax.legend(loc='upper center', bbox_to_anchor=(0.5, 1.2), ncol=3, fancybox=True, shadow=True)
plt.setp(ax.get_xticklabels(), rotation=45)
sns.move_legend(ax, "lower center", bbox_to_anchor=(.5, 1), ncol=3, title=None, frameon=False)

widthbars=len(ax.patches)*[0.2]
count=0
for bar,newwidth in zip(ax.patches,widthbars):
    x = bar.get_x()
    width = bar.get_width()
    count+=1
    
    if count<11:
        bar.set_width(newwidth)
    elif count<21:
        bar.set_width(newwidth)
        bar.set_x(x-newwidth/3)
    else:
        bar.set_width(newwidth)
        bar.set_x(x-2*newwidth/3)
plt.show()
```
