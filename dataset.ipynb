```python
def UNet_dataset_with_valid(train_ratio,test_ratio,valid_ratio): # 0.7,0.15,0.15
    
    LoadData_Info_without_normal()

    l=[]

    test_ratio  = 1 - train_ratio
    valid_ratio = valid_ratio/test_ratio # 0.15/0.3=0.5
    test_ratio  = 1 - valid_ratio
    
    for i in range(2):
        X_train,X_rem,y_train,y_rem=train_test_split(data[i][0], data[i][1], train_size=train_ratio, random_state=42)
        X_valid, X_test, y_valid, y_test=train_test_split(X_rem,y_rem, test_size=test_ratio,random_state=42)
        l.append([X_train, y_train, X_valid, y_valid, X_test, y_test])
    
        for i in range(6):
            l[-1][i]=np.array(l[-1][i],dtype=object)

    X_train  = []
    y_train  = []
    
    X_valid  = []
    y_valid  = []
    
    X_test   = []
    y_test   = []
    
#     for i in range(3):
#         X_train.append(l[i][0])
#         y_train.append(l[i][1])
        
#         X_valid.append(l[i][2])
#         y_valid.append(l[i][3])
        
#         X_test.append(l[i][4])
#         y_test.append(l[i][5])
        
    for i in range(2):
        X_train.append(l[i][0])
        y_train.append(l[i][1])
        
        X_valid.append(l[i][2])
        y_valid.append(l[i][3])
        
        X_test.append(l[i][4])
        y_test.append(l[i][5])

    
    X_train=np.concatenate(X_train, axis=0)
    X_test=np.concatenate(X_test, axis=0)
    X_valid=np.concatenate(X_valid, axis=0)

    y_train=np.concatenate(y_train, axis=0)
    y_test=np.concatenate(y_test, axis=0)
    y_valid=np.concatenate(y_valid, axis=0)
        
    X_train,y_train=shuffle(X_train, y_train, random_state=0)
    X_test,y_test=shuffle(X_test, y_test, random_state=0)
    X_valid,y_valid=shuffle(X_valid, y_valid, random_state=0)   
    
    X_train=X_train.astype(np.float32)
    y_train=y_train.astype(np.float32)

    X_test=X_test.astype(np.float32)
    y_test=y_test.astype(np.float32)

    X_valid=X_valid.astype(np.float32)
    y_valid=y_valid.astype(np.float32)

#     print(len(X_train),len(X_valid),len(X_test),len(y_train),len(y_valid),len(y_test))
    
    for i in range(len(X_train)):
        X_train[i]=X_train[i].astype(np.float32)
    for i in range(len(y_train)):
        y_train[i]=y_train[i].astype(np.float32)

    for i in range(len(X_test)):
        X_test[i]=X_test[i].astype(np.float32)
    for i in range(len(y_test)):
        y_test[i]=y_test[i].astype(np.float32)

    for i in range(len(X_valid)):
        X_valid[i]=X_valid[i].astype(np.float32)
    for i in range(len(y_valid)):
        y_valid[i]=y_valid[i].astype(np.float32)
    
    return [X_train, X_valid, X_test, y_train, y_valid, y_test]
```

```python
def UNet_dataset(train_ratio): # 0.7,0.3
    
#     LoadData_Info()
    LoadData_Info_without_normal()
    l=[]
    
#     for i in range(3):
    for i in range(2):
        X_train,X_test,y_train,y_test=train_test_split(data[i][0], data[i][1], train_size=train_ratio, random_state=42)
        l.append([X_train, y_train, X_test, y_test])
#         print("*********************************************")
        for i in range(4):
            l[-1][i]=np.array(l[-1][i],dtype=object)

    #X_train, X_test, y_train, y_test

    X_train = []
    y_train  = []

    X_test = []
    y_test = []
    
#     for i in range(3):
    for i in range(2):
        X_train.append(l[i][0])
        y_train.append(l[i][1])
        X_test.append(l[i][2])
        y_test.append(l[i][3])

    X_train=np.concatenate(X_train, axis=0)
    X_test=np.concatenate(X_test, axis=0)

    y_train=np.concatenate(y_train, axis=0)
    y_test=np.concatenate(y_test, axis=0)

#     return;
    
    X_train,y_train=shuffle(X_train, y_train, random_state=0)
    X_test,y_test=shuffle(X_test, y_test, random_state=0)
    
    X_train=X_train.astype(np.float32)
    y_train=y_train.astype(np.float32)

    X_test=X_test.astype(np.float32)
    y_test=y_test.astype(np.float32)

    for i in range(len(X_train)):
        X_train[i]=X_train[i].astype(np.float32)
    for i in range(len(y_train)):
        y_train[i]=y_train[i].astype(np.float32)

    for i in range(len(X_test)):
        X_test[i]=X_test[i].astype(np.float32)
    for i in range(len(y_test)):
        y_test[i]=y_test[i].astype(np.float32)
    
    return [X_train, X_test, y_train, y_test]
```

```python
def split_urls(split_ratio):
    
    LoadData_Info()
    
    benign_X,benign_y = shuffle(data_url[0][0], data_url[0][1], random_state=0)
    malignant_X,malignant_y = shuffle(data_url[1][0], data_url[1][1], random_state=0)
    normal_X,normal_y = shuffle(data_url[2][0], data_url[2][1], random_state=0)
    
    benign_X_train = benign_X[:int(len(benign_X)*split_ratio)]
    benign_X_test = benign_X[int(len(benign_X)*split_ratio):]    
    
    malignant_X_train = malignant_X[:int(len(malignant_X)*split_ratio)]
    malignant_X_test  = malignant_X[int(len(malignant_X)*split_ratio):]
    
    normal_X_train = normal_X[:int(len(normal_X)*split_ratio)]
    normal_X_test = normal_X[int(len(normal_X)*split_ratio):]
    
    benign_y_train = benign_y[:int(len(benign_y)*split_ratio)]
    benign_y_test = benign_y[int(len(benign_y)*split_ratio):]    
    
    malignant_y_train = malignant_y[:int(len(malignant_y)*split_ratio)]
    malignant_y_test  = malignant_y[int(len(malignant_y)*split_ratio):]    
    
    normal_y_train = normal_y[:int(len(normal_y)*split_ratio)]
    normal_y_test = normal_y[int(len(normal_y)*split_ratio):]
    
    return [(benign_X_train,benign_y_train),(benign_X_test,benign_y_test),(malignant_X_train,malignant_y_train),(malignant_X_test,malignant_y_test),(normal_X_train,normal_y_train),(normal_X_test,normal_y_test)]
```
